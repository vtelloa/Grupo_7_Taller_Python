I. Utilizando un “if else” cree una estructura que evalúe “x” tal que:

Si x es múltiplo de 2 imprima “Es múltiplo de 2”.

Si x es múltiplo de 3 imprima “Es múltiplo de 3”

Si x es múltiplo de 2 y 3 imprima “Es múltiplo de 2 y 3”

Para otro caso, imprime “Otro caso”

Muestra un ejemplo con x = 6.

```python
x = 6

if x % 2 == 0 and x % 3 == 0:
  print("Es múltiplo de 2 y 3")
elif x % 2 == 0:
  print("Es múltiplo de 2")
elif x % 3 == 0:
  print("Es múltiplo de 3")
else:
  print("Otro caso")
```

II. Utilizando un “for loop” y el condicional “if else” para un rango del 1 al 100, evalúe “x”
tal que:

Si x es múltiplo de 3 imprima “Fizz”

Si x es múltiplo de 5 imprima “Buzz”

Si x es múltiplo de 3 y 5 imprima “FizzBuzz”

Para el resto de los casos, solo imprima el número.

```python
for x in range(1, 101):
  if x % 3 == 0 and x % 5 == 0:
    print("FizzBuzz")
  elif x % 3 == 0:
    print("Fizz")
  elif x % 5 == 0:
    print("Buzz")
  else:
    print(x)
```

III. Crea una función calcular_descuento() que permita calcular el descuento aplicado a un
producto. Si el total es mayor o igual a 1000, aplica un 20% de descuento, si el total es
mayor o igual a 500 pero menor a 1000, aplica un 10% de descuento. Si el total es menor
a 500, no aplica ningún descuento. La función debe devolver el total final después de
aplicar el descuento. Prueba la función con un total de compra de 1200, 750 y 300.

```python
def calcular_descuento(total):
  if total >= 1000:
    total *= 0.8 # total = total*0.8
  elif total >= 500 and total < 1000:
    total *= 0.9 # total = total*0.9
  return total
```

```python
total = 1200
total_final = calcular_descuento(total)
print(total_final)
```

```python
total = 750
total_final = calcular_descuento(total)
print(total_final)
```

```python
total = 300
total_final = calcular_descuento(total)
print(total_final)
```

IV. Crea una función llamada clasificar_edad() que reciba como parámetro la edad de una
persona. La función debe clasificar a la persona en una de las siguientes categorías:

"Niño" si la edad es menor de 12 años.

"Adolescente" si la edad está entre 12 y 17 años.

“Joven” si la edad está entre 18 y 24 años.

"Adulto" si la edad está entre 25 y 64 años.

"Adulto Mayor" si la edad es mayor o igual a 65 años.

La función debe devolver la categoría correspondiente. Prueba la función con las
edades de 10, 19, 25 y 70.

```python
def clasificar_edad(edad):
  if edad >=0 and edad < 12:
    return "Niño"
  elif edad >=12 and edad <= 17:
    return "Adolescente"
  elif edad >=18 and edad <= 24:
    return "Joven"
  elif edad >=25 and edad <= 64:
    return "Adulto"
  elif edad < 0:
    return "Error: las edades no son negativas."
  else:
    return "Adulto Mayor"
```

```python
clasificar_edad(10)
```

```python
clasificar_edad(19)
```

```python
clasificar_edad(25)
```

```python
clasificar_edad(70)
```

```python
clasificar_edad(-5)
```

Crea una función calcular_craest(media_personal, media_curso, desviacion_curso)
que reciba como tres únicos inputs tu media, la media del curso y la desviación
estándar del curso y te devuelva tu craest calculado. No te olvides documentar la
función usando triple comillas “””.
Prueba la función con los datos de algún curso que hayas llevado.

```python
def calcular_craest(media_personal, media_curso, desviacion_curso):
  '''
  Esta es una función para calcular el CRAEST.
  media_personal: promedio de notas del estudiante
  media_curso: promedio de notas de todo el curso
  desviacion_curso: desviación estándar de las notas de todo el curso
  '''
  media_personal = media_personal
  media_curso = media_curso
  desviacion_curso = desviacion_curso
  craest = ((media_personal - media_curso)*10 / desviacion_curso) + 50
  return craest
```
