III. Con ambos dataframes, realiza un inner join. Guarda el resultado en un nuevo dataframe
llamado cuadro_1

```python
cuadro_1 = pd.merge(nfo,nombres, on='Ciudad', how = 'inner')

cuadro_1.head()
```

IV. Resume las estadísticas descriptivas del dataframe y responde ¿Cuál es la mínima
densidad poblacional? ¿A cuál ciudad corresponde? ¿Y la máxima?

```python
cuadro_1.describe().T
```

```python
# Encontrar la mínima y máxima densidad poblacional
min_densidad = cuadro_1['Densidad_poblacion_por_km2'].min()
max_densidad = cuadro_1['Densidad_poblacion_por_km2'].max()

# Encontrar las ciudades correspondientes
ciudad_min_densidad = cuadro_1.loc[cuadro_1['Densidad_poblacion_por_km2'] == min_densidad, 'Ciudad'].values[0]
ciudad_max_densidad = cuadro_1.loc[cuadro_1['Densidad_poblacion_por_km2'] == max_densidad, 'Ciudad'].values[0]

print(f'La mínima densidad poblacional es {min_densidad} por km2 y se encuentra en {ciudad_min_densidad}.')
print(f'\nLa máxima densidad poblacional es {max_densidad} por km2 y se encuentra en {ciudad_max_densidad}.')
```

V. Realizar un gráfico de barras donde se vea la cantidad de habitantes en cada ciudad

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.bar(cuadro_1['Ciudad'], cuadro_1['Habitantes'])
plt.xlabel('Ciudad')
plt.ylabel('Cantidad de Habitantes')
plt.title('Cantidad de Habitantes por Ciudad')
plt.xticks(rotation=45)  # Rota etiquetas
plt.show()
```

VI. Realizar un gráfico de dispersión entre la altura y el número de habitantes en las
ciudades del dataframe.

```python
plt.figure(figsize=(10, 6))
plt.scatter(cuadro_1['Altitud_m'], cuadro_1['Habitantes'])
plt.xlabel('Altura (m)')
plt.ylabel('Número de Habitantes')
plt.title('Relación entre Altura y Número de Habitantes del Perú')
plt.grid(True)  # cuadrícula
plt.show()
```
